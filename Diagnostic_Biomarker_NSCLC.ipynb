```python
# Import necessary libraries
import numpy as np
import pandas as pd
```

```python
import os
current_directory = os.getcwd()
print(current_directory) 
```

```python
# Load the GTEx expression data
gtex_expression = pd.read_csv(os.path.join(os.getcwd(), 'GTEx_Analysis/GTEx_Analysis_2017-06-05_v8_RNASeQCv1.1.9_gene_tpm.gct')
                              , sep='\t', skiprows=2)

gtex_expression
```

```python
# Load the GTEx metadata
gtex_metadata = pd.read_csv(os.path.join(os.getcwd(), 'GTEx_Analysis/GTEx_Analysis_v8_Annotations_SampleAttributesDS.txt')
                              , sep='\t')

gtex_metadata
```

```python
# Filter gtex_metadata based on the "SMTS" column being "Lung" (tissue type being lung)
sample_lung = gtex_metadata[gtex_metadata['SMTS'] == 'Lung']

sample_lung
```

```python
# Set Sample IDs as index
sample_lung.set_index('SAMPID', inplace=True)

# Convert the index (Sample ID) to a list
sample_lung_ids = sample_lung.index.tolist()

sample_lung_ids[:5]
```

```python
gtex_expression.set_index('Name', inplace=True)
```

```python
# Ensure that Sample IDs match by cleaning both lists (optional: lowercase and strip spaces)
sample_ids_cleaned = [x.strip().upper() for x in sample_lung_ids]  # Remove any extra spaces and make lowercase
gtex_expression.columns = [x.strip().upper() for x in gtex_expression.columns]  # Clean df2's column names the same way

sample_ids_cleaned[:5]
```

```python
gtex_expression.columns[:5]
```

```python
# Try filtering again
# Use the intersection of sample_ids_cleaned and df2's columns to avoid KeyErrors
matching_sample_ids = list(set(sample_ids_cleaned).intersection(set(gtex_expression.columns)))
len(matching_sample_ids)
```

```python
# Use the extracted Sample IDs to filter columns in gtex_expression, keeps only the columns that match the Sample IDs from gtex_expression
gtex_expression_lung = gtex_expression[matching_sample_ids]

gtex_expression_lung
```

```python
gtex_expression_lung.to_csv('GTEx_gene_tpm_lung.csv')
```

```python
# Filter out genes with low expression across all samples
# Set a threshold to keep only genes with a TPM value > 1 in at least 10% of samples
threshold = 1
min_samples = int(0.10 * len(gtex_expression_lung.columns))
gtex_expression_filtered = gtex_expression_lung[gtex_expression_lung.gt(threshold).sum(axis=1) > min_samples]

gtex_expression_filtered
```
